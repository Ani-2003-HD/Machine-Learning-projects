```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#let's take the dataset present in the sklearn library
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
```

```python
#load the data to a dataframe
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df.head()
```

```python
#scaling the data
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(df)
scaled_data = scale.transform(df)
```

```python
#importing and training the pca(Principle component analysis) model
from sklearn.decomposition import PCA
model = PCA(n_components=2)
model.fit(scaled_data)
```

```python
X_pca = model.transform(scaled_data)
print(scaled_data.shape)
X_pca.shape
```

```python
#let's plot the reduced to 2 dimension data
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=cancer['target'], cmap='plasma')
plt.xlabel('First PC')
plt.ylabel('Second PC')
```

```python
model.components_
```

```python
#let's see a heatmap
df_comp = pd.DataFrame(model.components_, columns=cancer['feature_names'])
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
