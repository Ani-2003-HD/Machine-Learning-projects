```python
#importing the required libraries and dataset and the model also
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn import svm
```

```python
#lets create 40 seperable points with 2 centers
X, Y = make_blobs(n_samples=40, centers=2, random_state=20)
```

```python
#training the svm model on the generated data
model = svm.SVC(kernel='linear', C=1)
model.fit(X, Y)
```

```python
#let's see the data in a graph
plt.scatter(X[:,0], X[:,1], c=Y, s=30, cmap=plt.cm.Paired)
plt.show()
```

```python
#let's predict the result for a new data(we give the data manually rather than creating test data)
new_data = [[3,4],[5,6]]
model.predict(new_data)
```

```python
#let's see what's going on behind the scene
model = svm.SVC(kernel='linear', C=1)
model.fit(X, Y)
plt.scatter(X[:,0], X[:,1], c=Y, s=30, cmap=plt.cm.Paired)

#plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

#creating a grid to evalualte the model
xx = np.linspace(xlim[0],xlim[1],30)
yy = np.linspace(ylim[0],ylim[1],30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = model.decision_function(xy).reshape(XX.shape)

#plotting decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels =[-1,0,1], alpha=0.5, linestyles=['--','-','--'])

#plot the support vectors
ax.scatter(model.support_vectors_[:,0], model.support_vectors_[:,0], s=100, linewidth=1, facecolors='none')
plt.show()
```
