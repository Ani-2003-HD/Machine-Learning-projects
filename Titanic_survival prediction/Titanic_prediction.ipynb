```python
#import the necessary libraries
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train.head()
```

```python
print(train.info())
test.info()
```

```python
#let's drop the cabin column as it contains about 70% null values
train.drop(columns=['Cabin'], inplace=True)
test.drop(columns=['Cabin'], inplace=True)

#let's check for null values in other columns
train.isnull().sum()
```

```python
train['Embarked'].value_counts()
```

```python
train['Embarked'].fillna('S', inplace=True)
train.isnull().sum()
```

```python
test['Fare'].fillna(test['Fare'].mean(), inplace=True)
test.isnull().sum()
```

```python
train_age = np.random.randint(train['Age'].mean() - train['Age'].std(), train['Age'].mean() + train['Age'].std(), 177)
train_age
```

```python
test_age = np.random.randint(test['Age'].mean() - test['Age'].std(), test['Age'].mean() + test['Age'].std(), 86)
test_age
```

```python
train['Age'][train['Age'].isnull()] = train_age
```

```python
test['Age'][test['Age'].isnull()] = test_age
```

```python
print(train.isnull().sum())
train.isnull().sum()
```

```python
def calculate(n):
    if n==1: return "Alone"
    elif n>1 and n<5: return "Medium"
    else: return "Large"

train['family'] = train['SibSp'] + train['Parch'] + 1
train['fam_size'] = train['family'].apply(calculate)
train.head()
```

```python
test['family'] = test['SibSp'] + test['Parch'] + 1
test['fam_size'] = test['family'].apply(calculate)
```

```python
train = train.drop(columns=['SibSp', 'Parch', 'family', 'Name', 'Fare', 'PassengerId', 'Ticket'], axis=1)
train.head()
```

```python
test = test.drop(columns=['SibSp', 'Parch', 'family', 'Name', 'Fare', 'Ticket'], axis=1)
test.head()
```

```python
train = pd.get_dummies(train, columns=['Pclass', 'Sex', 'Embarked', 'fam_size'], drop_first=True)
test = pd.get_dummies(test, columns=['Pclass', 'Sex', 'Embarked', 'fam_size'], drop_first=True)
train.head()
```

```python
X = train.iloc[:,1:]
Y = train.iloc[:,0]
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=15)
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
DTC.fit(X_train,Y_train)
```

```python
Y_DTC = DTC.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(X_train,Y_train)
```

```python
Y_RFC = RFC.predict(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(X_train,Y_train)
```

```python
Y_GNB = GNB.predict(X_test)
```

```python
from sklearn.svm import SVC
SV = SVC()
SV.fit(X_train,Y_train)
```

```python
Y_SVC = SV.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Decision Tree Classifier :")
print("Accuracy", accuracy_score(Y_test,Y_DTC))
print("    Classification report:")
print(classification_report(Y_test,Y_DTC))
print("    Confusion matrix")
print(confusion_matrix(Y_test,Y_DTC))
```

```python
print("Random Forest Classifier :")
print("Accuracy", accuracy_score(Y_test,Y_RFC))
print("    Classification report:")
print(classification_report(Y_test,Y_RFC))
print("    Confusion matrix")
print(confusion_matrix(Y_test,Y_RFC))
```

```python
print("Naive Bayes Classifier :")
print("Accuracy", accuracy_score(Y_test,Y_GNB))
print("    Classification report:")
print(classification_report(Y_test,Y_GNB))
print("    Confusion matrix")
print(confusion_matrix(Y_test,Y_GNB))
```

```python
print("Support Vector Classifier :")
print("Accuracy", accuracy_score(Y_test,Y_SVC))
print("    Classification report:")
print(classification_report(Y_test,Y_SVC))
print("    Confusion matrix")
print(confusion_matrix(Y_test,Y_SVC))
```
